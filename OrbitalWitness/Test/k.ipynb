```python
file = '/Users/talib/Documents/Projects/OrbitalWitness/Official_Copy_Register_EGL363613.pdf'
data= tb.read_pdf(file, pages = '5', columns = [200, 265, 300, 320], pandas_options={'header': None}, stream=True)
```

```python
# datadf = np.array(data)
datas= np.array(data, order='C')
# datas.resize((1,55,5))
df = pd.DataFrame(datas[0])
df['border'] = df.apply(lambda x: 1 if re.findall('^[0-9].*[0-9]$', str(x[0])) else 0, axis = 1)
df['row'] = df['border'].transform('cumsum')
# print(df)
schedule_formatted = pd.DataFrame()
schedule = df[[0,1, 2, 3, 4, 'row']]
schedule = schedule[schedule[1].notnull()]
# schedule['index'] = schedule.groupby('row').agg({'row': list})
# display(schedule)
schedule['index'] = schedule.groupby('row').cumcount()+1
schedule = schedule[schedule['row']!=0]
schedule = schedule.replace(np.NaN,'')
schedule = schedule.pivot_table(index = ['row'], columns = ['index'], values = [1,2,3,4],aggfunc=lambda x: ' '.join(x)).reset_index()
schedule = schedule.replace(np.NaN,'')
schedule_formatted['Registration date and plan ref.'] = schedule.iloc[0:,0:6].apply(lambda x: ''.join(map(str, x)), axis=1)
schedule_formatted['Property description'] = schedule.iloc[0:,7:12].apply(lambda x: ''.join(map(str, x)), axis=1)
schedule_formatted['Date of lease and term.'] = schedule.iloc[0:,13:18].apply(lambda x: ''.join(map(str, x)), axis=1)
schedule_formatted['Lessee''s title.'] = schedule.iloc[0:,19:24].apply(lambda x: ''.join(map(str, x)), axis=1)
display(schedule_formatted)
```

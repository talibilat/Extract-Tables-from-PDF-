```python
import json
import pandas as pd

# Load the JSON data
with open('/Users/talib/Documents/Projects/OrbitalWitness/schedule_of_notices_of_lease_examples.json', 'r') as file:
    data = json.load(file)

# Initialize an empty list to store the structured data
structured_data = []

# Parse each entry and add to the structured data list
for entry in data[0]["leaseschedule"]["scheduleEntry"]:
    lines = entry['entryText']
    registration_date_and_plan_ref = lines[0]
    property_description = []
    date_of_lease_and_term = []
    lessees_title = ""
    notes = []

    # Split the entry into columns and notes
    for line in lines[1:]:
        
        if line.startswith("NOTE:"):
            notes.append(line)
        else:
            if len(property_description) == 0:
                property_description.append(line)
            elif len(date_of_lease_and_term) == 0:
                date_of_lease_and_term.append(line)
            else:
                lessees_title = line

    # Combine the parsed information
    parsed_entry = {
        "Registration date and plan ref": registration_date_and_plan_ref,
        "Property description": ' '.join(property_description),
        "Date of lease and term": ' '.join(date_of_lease_and_term),
        "Lessee's title": lessees_title,
        "Notes": ' '.join(notes)
    }

    structured_data.append(parsed_entry)

# Convert the structured data to a DataFrame
df = pd.DataFrame(structured_data)

# Save the DataFrame to a CSV file
df.to_csv('structured_schedule_of_notices.csv', index=False)

print("Data has been structured and saved to 'structured_schedule_of_notices.csv'.")
```
